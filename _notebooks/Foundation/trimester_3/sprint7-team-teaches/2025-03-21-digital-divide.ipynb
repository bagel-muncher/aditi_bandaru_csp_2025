```python
%%js
function exploreDigitalDivide() {
    // Select what data to display
    const metric = prompt(
      "Which digital divide metric would you like to explore?\n" +
      "1: Income-based internet access\n" +
      "2: Urban vs rural access\n" +
      "3: Age-based internet usage\n" +
      "4: Global connectivity rates"
    );
  
    // Sample data
    const incomeData = {
      categories: ["Lowest 20%", "Second 20%", "Middle 20%", "Fourth 20%", "Highest 20%"],
      values: [62, 71, 80, 88, 95]
    };
  
    const locationData = {
      categories: ["Urban", "Rural"],
      values: [94, 83]
    };
  
    const ageData = {
      categories: ["18-29", "30-49", "50-64", "65+"],
      values: [97, 93, 88, 61]
    };
  
    const globalData = {
      categories: ["North America", "Europe", "Asia Pacific", "Latin America", "Middle East/Africa"],
      values: [90, 87, 54, 68, 40]
    };
  
    // Select dataset based on user choice
    let data;
    let title;
  
    switch(metric) {
      case "1":
        data = incomeData;
        title = "Internet Access by Income Level (%)";
        break;
      case "2":
        data = locationData;
        title = "Urban vs Rural Internet Access (%)";
        break;
      case "3":
        data = ageData;
        title = "Internet Usage by Age Group (%)";
        break;
      case "4":
        data = globalData;
        title = "Internet Access by Region (%)";
        break;
      default:
        alert("Invalid selection!");
        return;
    }
  
    // Display the data (in a real app, this would create a chart)
    console.log(title);
    console.log("=".repeat(title.length));
  
    const maxValue = Math.max(...data.values);
    const maxBarLength = 40;
  
    for (let i = 0; i < data.categories.length; i++) {
      const barLength = Math.round((data.values[i] / maxValue) * maxBarLength);
      const bar = "█".repeat(barLength);
      console.log(`${data.categories[i].padEnd(12)}: ${bar} ${data.values[i]}%`);
    }
  }
  
  // Run this function to explore different aspects of the digital divide
  
exploreDigitalDivide();
```

## Essay Practice Questions

2. **Compare and contrast the digital divide challenges faced in urban versus rural communities. What solutions might be effective in each context?**

In urban areas, challenges often stem from economic and social inequalities rather than infrastructure limitations. Many low-income families cannot afford high-speed internet or personal devices, and some individuals may lack the digital literacy skills needed to take full advantage of online resources. Additionally, public spaces like libraries or community centers may be overcrowded, limiting access for those who rely on shared resources.

In contrast, rural communities face more significant infrastructure barriers, such as the lack of broadband internet or cellular service. Even if households can afford internet access, providers may not offer high-speed connections in remote areas, making it difficult for students and workers to engage in online learning or remote work. Additionally, rural communities often have fewer public access points, such as libraries with internet services, further limiting connectivity options.

Effective solutions for urban areas could include increasing affordability programs for low-income households—such as government-subsidized broadband services—can help bridge the gap. Expanding digital literacy training in schools and community centers can also empower individuals to use technology more effectively. In rural areas, investment in broadband infrastructure, such as expanding fiber-optic networks and deploying satellite internet services, is crucial. Additionally, mobile Wi-Fi hotspots and community-driven initiatives, such as local internet cooperatives, can provide alternative access to underserved populations.

3. **Analyze how the COVID-19 pandemic exposed and potentially exacerbated existing digital divides. Include at least three specific examples.**

The COVID-19 pandemic highlighted many disparities in access to technology and the internet. One major issue was the unequal access to remote learning. Many students, especially in low-income and rural communities, lacked reliable internet or personal devices, making it difficult to attend virtual classes. For example, some students had to complete assignments on smartphones or share a single computer among multiple siblings, leading to decreased engagement and learning loss.

Another example is the impact on remote work and employment opportunities. While many professionals transitioned to working from home, individuals in jobs requiring physical presence—often lower-income workers—had fewer options. Additionally, those without reliable internet connections faced challenges in applying for jobs, attending virtual interviews, or accessing online job training programs, deepening economic inequalities.

The pandemic also highlighted disparities in access to healthcare services. Telehealth became a crucial tool for medical consultations, but many elderly individuals or those in rural areas struggled to access virtual healthcare due to poor internet connectivity or lack of familiarity with digital platforms. As a result, some individuals faced delays in receiving medical care, further exacerbating health inequalities.
